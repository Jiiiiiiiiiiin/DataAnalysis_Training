```python
mpg1 = sns.load_dataset('mpg')
mpg1.tail()
```

```python
mpg1['hp_per_cc'] = mpg1['displacement'] / mpg1['horsepower']
mpg1.tail()
```

```python
mpg1['manufacturer'] = mpg1.name.apply(lambda x: x.split()[0])
```

```python
mpg1['model'] = mpg1.name.apply(lambda x: ' '.join(x.split()[1:]))
```

```python
del mpg1['name']
mpg1.tail()
```

```python
mpg1.groupby('cylinders')['mpg'].mean()
```

```python
mpg1.groupby('origin')['hp_per_cc'].mean()
```

```python
new = pd.DataFrame(mpg1.groupby('manufacturer')['model'].count())
new.head()
```

```python
new['mpg'] = pd.DataFrame(mpg1.groupby('manufacturer')['mpg'].mean())
new.head()
```

```python
new['five']  = new.apply(lambda x: x.model if x.model >= 5 else 'na', axis=1)
new.head()
```

```python
new.sort_values(['mpg'], ascending=False)
```
