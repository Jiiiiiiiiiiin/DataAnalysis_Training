```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='NanumBarunGothic')
```

## Figure 객체
 
 * 자동시작
 * 모든 그림은 Figure 객체. 정식으로는 matplotlib.figure.Figure 클래스 객체에 포함되어 있다. 내부 플롯(inline plot)이 아닌 경우에는 하나의 Figure는 하나의 아이디 숫자와 윈도우(Window)를 가진다. 주피터 노트북에서는 윈도우 객체가 생성되지 않지만 파이썬을 독립 실행하는 경우에는 하나의 Figure당 하나의 윈도우를 별도로 가진다.
* 하나의 figure 안에 하나 이상의 axes 객체를 가진다
* Figure 안에 Axes를 생성하려면 원래 subplot 명령을 사용하여 명시적으로 Axes 객체를 얻어야 한다. 그러나 plot 명령을 바로 사용해도 자동으로 Axes를 생성해 준다.


```python
np.random.seed(2021)
f1 = plt.figure(figsize=(10, 2))
plt.title("figure size : (10, 2)")
plt.plot(np.random.randn(100))
plt.grid()
plt.show()
```

## Axes 객체와 subplot

* subplots(행 숫자, 열 숫자) subplot(행 숫자, 열 숫자, 순서)
* subplot은 axes객체가 됨

```python
x1 = np.linspace(0.0, 5.0)     
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
ax1 = plt.subplot(2, 1, 1)       # Axes
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
plt.grid()

ax2 = plt.subplot(2, 1, 2)       # Axes
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
plt.grid()

plt.tight_layout()   #plot간 간격조정
plt.show()
```

```python
np.random.seed(2021)

plt.subplot(221)   #(2,2,1)=221
plt.plot(np.random.rand(5))
plt.title("axes 1")
plt.grid()

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("axes 2")
plt.grid()

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("axes 3")
plt.grid()

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")
plt.grid()

plt.tight_layout()
plt.show()
```

```python
_, axes = plt.subplots(2, 2)  #fig 는 _ 숨김. subplots 2x2 array를 만들겠다

np.random.seed(201)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 0].grid()
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[0, 1].grid()
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 0].grid()
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")
axes[1, 1].grid()

plt.tight_layout()
plt.show()
```

```python
np.random.seed(2021)
_, axes = plt.subplots(2,2)

for i in range(4):
  ax = axes[i//2, i%2] #행 인덱스 i//2, 열인덱스 i%2
  ax.plot(np.random.rand(5))
  ax.set_title(f'axes {i+1}')
  ax.grid()

plt.tight_layout()
plt.show()
```

```python
np.random.seed(2021)
fig, axes = plt.subplots(2,2, figsize=(6,4))      #fig 를 _하면 안쓰겠다는의미. 모든 함수 적용

for i in range(4):
  ax = axes[i//2, i%2]
  ax.plot(np.random.rand(5))
  ax.set_title(f'axes {i+1}')
  ax.set_xlabel('x축'); ax.set_ylabel('y축')
  ax.grid()

fig.suptitle('4개의 그림', fontsize=14)
plt.tight_layout()
plt.show()
```
